### 1.2 Polynomial regression (10 points)

Write a method `fit_polynomial(x, t, M)` that finds the maximum-likelihood solution of an _unregularized_ $M$-th order polynomial for some dataset `x`. The error function to minimize w.r.t. $\bw$ is:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt)$

where $\bPhi$ is the _feature matrix_ (or _design matrix_) as explained in Bishop's book at section 3.1.1, $\bt$ is the vector of target values. Your method should return a vector $\bw$ with the maximum-likelihood parameter estimates, as well as the _feature matrix_ $\bPhi$.

```python
def designmatrix(x, M):
    return np.array([x ** p for p in range(M + 1)]).T

def fit_polynomial(x, t, M):
    Phi = designmatrix(x, M)
    w_ml = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ t
    
    return w_ml, Phi
```

a) (10 points) Write a method `fit_polynomial_reg(x, t, M, lamb)` that fits a _regularized_ $M$-th order polynomial to the periodic data, as discussed in the lectures, where `lamb` is the regularization term _lambda_. (Note that 'lambda' cannot be used as a variable name in Python since it has a special meaning). The error function to minimize w.r.t. $\bw$:

$E(\bw) = \frac{1}{2} (\bPhi\bw - \bt)^T(\bPhi\bw - \bt) + \frac{\lambda}{2} \mathbf{w}^T \mathbf{w}$

The function should return $\bw$ and $\bPhi$.

```python
def fit_polynomial_reg(x, t, m, lamb):
    Phi = designmatrix(x, m)
    w_ml = np.linalg.inv(lamb * np.eye(m + 1) + Phi.T @ Phi) @ Phi.T @ t
    
    return w_ml, Phi
```

a) (5 points) First of all, write a method `pred_error(x_train, x_valid, t_train, t_valid, M, lamb)` that compares the prediction of your method `fit_polynomial_reg` for a given set of parameters $M$ and $\lambda$ to `t_valid`. It should return the prediction error for a single fold.

```python
def pred_error(x_train, x_valid, t_train, t_valid, M, reg):
    # estimate w using ML approach from train data
    w_ml, _ = fit_polynomial_reg(x_train, t_train, M, reg)

    # calculate Phi for validation set
    Phi = designmatrix(x_valid, M)

    # predict output for validation set
    y_pred = Phi @ w_ml

    # calculate prediction error (MSE)
    pred_err = np.sum((y_pred - t_valid) ** 2) / len(x_valid)
    
    return pred_err
```

```python
def kfold_indices(N, k):
    all_indices = np.arange(N,dtype=int)
    np.random.shuffle(all_indices)
    idx = [int(i) for i in np.floor(np.linspace(0,N,k+1))]
    train_folds = []
    valid_folds = []
    for fold in range(k):
        valid_indices = all_indices[idx[fold]:idx[fold+1]]
        valid_folds.append(valid_indices)
        train_folds.append(np.setdiff1d(all_indices, valid_indices))
    return train_folds, valid_folds
```

```python
import itertools

def find_best_m_and_lamb(x, t):
    grid = list(
        itertools.product(
            np.linspace(0, 10, num=11, dtype=np.int), # M values
            np.logspace(-10, 0, num=11, base=np.e)    # lambda values
        )
    )

    cv_errors = []
    for M, lamb in grid:
        error = []

        # Iterate over train/validation splits
        train_folds, valid_folds = kfold_indices(10, 5)
        for train_idx, valid_idx in zip(train_folds, valid_folds):
            x_train = x[train_idx]
            x_valid = x[valid_idx]
            t_train = t[train_idx]
            t_valid = t[valid_idx]
            
            error.append( pred_error(x_train, x_valid, t_train, t_valid, M, lamb) )

        # Save error
        cv_errors.append(np.mean(error))

    return grid[np.argmin(cv_errors)] #M_best, lamb_best
```

### 1.6 Why grid search? (5 points)

Grid search is an commonly-used technique to tune hyper-parameters in a model.
Considering the case described in the previous step of this assignment, running a grid search over the possible parameter values (10 possible values for both $M$ and $\lambda$), results in two nested loops exploring $10 \times 10 = 100$ different configurations for the model. 

a) (3 points) Why do we want to optimize by changing the two hyperparameters at the same time, and not in a sequential way? We could initialise all parameters randomly, fix one parameter at a time and iterate over the other, resulting in only $10 + 10 = 20$ experiments!

As we are trying to find the pair of hyperparameters that minimizes the error function, we need to extensively search in the hyperparameter plane for the global minumum with respect to those parameters. If we were to choose our parameters in a sequential way, then we could end up with a scenario where, even though we did minimize the error w.r.t. the first parameter, the best option for the second parameter would still not result in the smallest attainable error. This is because we are only taking a partial derivative into consideration at a time, instead of the gradient for both (or all, if we have more than 2 hyperparameters) which would show us the true direction of the minimum.

For more complex models, the number of combinations easily explodes with the number of parameters. For example, with 5 parameters we would run $10 \times 10 \times 10 \times 10 \times 10 = 100,000$ experiments.

b) (2 points) Try to think or find in literature one alternative to grid search to tune hyper-parameters more efficiently. Explain very briefly (2-3 lines max) how this method avoids the combinatorial explosion we have see in grid search.

In order to avoid exploring the cartesian product of the possible hyperparameters, we can use an evolutionary optimization approach. In evoluationary optimization, we start with a set of randomly drawn hyperparameters $ \{ M_i, \lambda_i \}_{i=1}^{N} $, which is the original **population**, and evaluate the performance of our model for each pair (in our case: compute the validation error). Then, we rank the pairs $ (M_i, \lambda_i) $, or tuples of hyperparameters in general, according to their performance, and replace the worst performing ones with new pairs using the operations of **mutation** (randomly altering the value of a parameter) and **cross-over** (mixing well performing pairs), resulting in a new population. We repeat this procedure until the error obtained by the top performing pair stops decreasing.

### 2.2 Compute Posterior (15 points)

You're going to implement a Bayesian linear regression model, and fit it to the periodic data. Your regression model has a zero-mean isotropic Gaussian prior over the parameters, governed by a single (scalar) precision parameter $\alpha$, i.e.:

$$p(\bw \;|\; \alpha) = \mathcal{N}(\bw \;|\; 0, \alpha^{-1} \bI)$$

The covariance and mean of the posterior are given by:

$$\bS_N= \left( \alpha \bI + \beta \bPhi^T \bPhi \right)^{-1} $$
$$\bm_N = \beta\; \bS_N \bPhi^T \bt$$

where $\alpha$ is the precision of the predictive distribution, and $\beta$ is the noise precision. 
See MLPR chapter 3.3 for background.

Write a method `fit_polynomial_bayes(x, t, M, alpha, beta)` that returns the mean $\bm_N$ and covariance $\bS_N$ of the posterior for a $M$-th order polynomial. In addition it should return the design matrix $\bPhi$. The arguments `x`, `t` and `M` have the same meaning as in question 1.2.

```python
def fit_polynomial_bayes(x, t, M, alpha, beta):
    Phi = designmatrix(x, M)

    S = np.linalg.inv(alpha * np.eye(M + 1) + beta * Phi.T @ Phi)
    m = beta * S @ Phi.T @ t
    
    return m, S, Phi
```

### 2.3 Prediction (10 points)

The predictive distribution of Bayesian linear regression is:

$$ p(t \;|\; \bx, \bt, \alpha, \beta) = \mathcal{N}(t \;|\; \bm_N^T \phi(\bx), \sigma_N^2(\bx))$$

$$ \sigma_N^2 = \frac{1}{\beta} + \phi(\bx)^T \bS_N \phi(\bx) $$

where $\phi(\bx)$ are the computed features for a new datapoint $\bx$, and $t$ is the predicted variable for datapoint $\bx$. 

Write a function that `predict_polynomial_bayes(x, m, S, beta)` that returns the predictive mean, variance and design matrix $\bPhi$ given a new datapoint `x`, posterior mean `m`, posterior variance `S` and a choice of model variance `beta`.

```python
def predict_polynomial_bayes(x, m, S, beta):
    Phi = designmatrix(x, m.shape[0]-1)

    # vectorized formulas for predictive mean & variance
    mean = Phi @ m
    sigma = 1 / beta + np.diag(Phi @ S @ Phi.T)
    
    return mean, sigma, Phi
```

### 2.5 Additional questions (10 points)

a) (5 points) Why is $\beta=16$ the best choice of $\beta$ in section 2.4?



By definition, $ \beta $ is the inverse variance of the noise of our datapoints (Bishop, Ch 1.2.5).

Since the datapoints used in Section 2.4 are generated by **gen_sine2**, which adds a Gaussian noise with variance $ \sigma^2 = 0.25^2 = \frac{1}{16}$, the best choice of $ \beta $ is also 16.

Normally, we do not have such an insight as to how noisy our dataset is, which makes this hyperparamter selection a separate task that needs to be dealt with. However, in this case we know the specific distribution that generated the dataset, so the most optimal performance will be achieved when using the same variance for our likelihood function.

b) (5 points) What problems do we face when it comes to choosing basis functions in linear models?

By applying basis functions in linear models, we are essentially allowing the algorithm to map each sample point to the predicted target value in a non linear way. The key thought behind the selection proccess is to transform our input in a way that corresponds to the problem we are trying to solve - i.e. modelling a quadratic function using a second order polynomial basis, or a sinusoidal function using a Fourier basis.

In real problems, however, the data samples have a much higher dimensionality than 2D or 3D, which makes visualization difficult. As a result, it is not trivial to determine the proper form of basis functions that we should use, which can have significant implications for the performance of our model. This can be observed both as underfitting, when our selection actually hinders the ability of the model to map the transformed input to the target values, but also overfitting, when using more basis functions than what is needed without any form of regularization. An intuitive example for the latter case can be seen in the plots for Section 1.3, where for $M=8$, our polynomial fit is nowhere close to the target function, even though the error appears to be zero for our training samples.

In addition, most of the common basis functions suffer from various shortcomings when used to solve a problem that they weren't designed for. For instance, polynomial functions are global functions of the input variable, and hence a change in one region can affect other regions. On the other hand, localized basis functions, such as Gaussians, tend to be overly confident in their predictions when extrapolating, which is often an undesired behaviour.
